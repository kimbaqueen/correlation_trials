```python
from scipy.stats import pearsonr
```

```python
day = absentee_data['Day of the week']
reason = absentee_data['Reason for absence']
month = absentee_data['Month of absence']
age = absentee_data['Age']
discipline_failure = absentee_data['Disciplinary failure']
service_time = absentee_data['Service time']
trans_expense = absentee_data['Transportation expense']
distance = absentee_data['Distance from Residence to Work']
pet = absentee_data['Pet']
hours = absentee_data['Absenteeism time in hours']
height = absentee_data['Height']
bmi = absentee_data['Body mass index']
hit = absentee_data['Hit target']
```

# Correlation Using Scipy PearsonR

```python
pearsonr_coefficient, p_value = pearsonr(day, reason)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(month, age)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(hit, discipline_failure)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(service_time, reason)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(trans_expense, distance)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(age, pet)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(hours, day)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(month, hours)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(bmi, height)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(pet, discipline_failure)
print('PearsonR Correlation Coefficient %0.3f'% (pearsonr_coefficient))
```
